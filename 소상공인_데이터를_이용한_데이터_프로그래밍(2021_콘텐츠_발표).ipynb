```python
# info는 데이터의 종류를 기반으로 번호, 열, 비어있지 않은 데이터의 값, 데이터 타입을 요약해 보여줌
# 용량도 확인가능
# import csv로 불러왔을 때와는 다르게 Dtype이 전부 문자열이 아니라 사람이 인식할 수 있는 데이터와 유사함을 알 수 있음

df.info()
```

```python
# 데이터 프로그래밍과에서는 데이터가 의미하는 정확한 뜻을 파악하고 이 데이터가 양질의 데이터인지 부족함이 있는 데이터인지 파악하는 것이 필요
# 공부를 위해서는 좋은 자료가 필요할까 그저 그런 자료가 필요할까?
# 빈칸이 많은 종류의 데이터를 활용할 수 있을까?

print(df.isnull().sum()) #빈칸의 개수를 탐색하여 더해주고 이를 출력함 
```

```python
#판다스로 읽어온 공공데이터에서 특정 조건을 만족하는 데이터만 뽑아내기

flag = df['시군구명'] =='춘천시'
new_df = df[flag]
new_df
```

```python
# 우리가 관심 있는 데이터의 종류를 추출하고
# 이를 기반으로 새로운 데이터를 만들어줌 

columns = ['상호명','상권업종대분류명','상권업종중분류명','상권업종소분류명','시도명','시군구명','도로명주소', '위도', '경도']
print(df.shape)
df = df[columns].copy() #df에 특정 열 값들만 복사하여 새롭게 설정해줌
df.shape
```

```python
#변화된 df 확인

df.info()
```

```python
#고유 값(중복되지 않은 값을 확인)

print(df['상권업종대분류명'].unique())
print()
print(df['상권업종중분류명'].unique())
print()
print(df['상권업종소분류명'].unique())
print()
print(df['시군구명'].unique())
```

```python
# 우리가 뽑아내고 싶은 조건을 만들어 준 후 
# 그 조건을 적용하여 새로운 판다스 df를 만들어주자!!!

flag1 = df['시군구명']=='춘천시'
flag2 = df['상권업종소분류명'] == '한식/백반/한정식'
my_df = df[flag1 & flag2].copy()
print(my_df.info())

# 지도 시각화를 위해 필요한 값만 추출한 df를 csv파일로 만들어줌

my_df.to_csv("/content/drive/MyDrive/데이터 프로그래밍/춘천 한식.csv", index=False,encoding='utf-8-sig') #index는 밑에 밑에 있는 0,1,2,3,4 -> 파일로 만들어줄 때는 이를 파일에 포함되게 만들어주지 않음
```
